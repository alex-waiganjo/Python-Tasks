```python
# SETS

names = {"John","Kelvin","Kamau","Jane","Dominick","Christine","Job"}
names
```

```python
for name in names:
    print(name)

type(names)
```

```python
# Accessing Items in a Set
print("Janee" not in names)
```

```python
# Add Items to Sets
names.add("Harmonize")
names
```

```python
# Remove an Item From a Set
to_list = list(names)
# to_list.remove("Harmonize")
# names = set(to_list)
to_list
```

```python
# Combine two sets into one
tz_names = {"Diamond","Alikiba","Jay","Mbosso","Aslay","Mavoko"}
# names.update(tz_names)
tz_names.update(to_list)
tz_names
```

```python
# Directly Remove an Element from a Set
# tz_names.remove("Alikiba")
# tz_names
tz_names.discard("Mbosso")
tz_names
```

```python
# tz_names.remove("Mavoko")
# tz_names
tz_names.remove("Bien")
tz_names
```

```python
# Empty the Set
tz_names.clear()
tz_names
```

```python
# Delete the Set Completely
del tz_names
tz_names
```

```python
# Joining Sets

# Intersection - Keeps the Duplicates Only
laptop_brand_a = {"HP","Lenovo","Toshiba","Dell"}
laptop_brand_b = {"HP","Lenovo","Toshiba","Dell","Acer"}
laptop_brand_a.intersection(laptop_brand_b)
laptop_brand_a
```

```python
# Difference  - Keeps Items in the first set that are not in the second Set
tv_brand_a = {"TCL","Sony","Samsung","Skyworth" }
tv_brand_b= {"Gratewall","Mooka","Vitron","Skyworth"}
get_diff = tv_brand_a.difference(tv_brand_b)
get_diff
```

```python
# Difference  - Keeps Items not present in both sets
tv_brand_a = {"TCL","Sony","Samsung","Skyworth" }
tv_brand_b= {"Gratewall","Mooka","Vitron","Skyworth"}
get_diff = tv_brand_a.symmetric_difference(tv_brand_b)
get_diff
```
