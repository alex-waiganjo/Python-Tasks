```python
import requests
import pandas as pd
```

```python
url = "https://dummyjson.com/carts"
cart_data = requests.get(url)
cart_data.status_code
```

```python
data = cart_data.json()
data
```

```python
all_data = data["carts"]
all_data
```

```python
len(all_data)
```

```python
df = pd.json_normalize(all_data)
df
```

```python
df.to_csv("carts.csv", index=False)
```
