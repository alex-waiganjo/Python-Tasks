```python
# Fetch data from the 
import pandas as pd
import requests

url = "https://raw.githubusercontent.com/odhiambokevin/luxdev_staff_python_project/refs/heads/main/staff_data.json"
data = requests.get(url)
print(data)
```

```python
emp_data = data.json()
emp_data
```

```python
len(emp_data) # Get total Items
```

```python
first_5_emps = emp_data[:5] # Get first 5 Users
first_5_emps
```

```python
# Use Pandas
df = pd.DataFrame(first_5_emps)
df
```

```python
df.to_csv("first_5_staffs.csv", index=False)
```
