```python
# Function to convert price(ksh) to usd
comp_price = [200, 350, 200, 150, 220]

def usd_to_ksh(comp_price, ksh_usd_rate=129):
   usd_price =  [price *ksh_usd_rate for price in comp_price]
   return usd_price

usd_to_ksh(comp_price)
```

```python
comp_price = [200, 350, 200, 150, 220]

def usd_to_ksh_part2(comp_price):
    empty_list = []
    for price in range(len(comp_price)):
        empty_list.append(comp_price[price]*129)
    return empty_list
    
usd_to_ksh_part2(comp_price)
```

```python
# Function to convert names to Uppercase
names = ["alex","jane","kev","Obinna"]

def names_to_upper(names):
    upper_names = [name.upper() for name in names]
    return upper_names

names_to_upper(names)
```

```python
# Function to square a given list
mylist = [2,4,6,8]

def square_list(mylist):
    p = [x**2 for x in mylist]
    return p

square_list(mylist)
```
