```python
"""
Tuples
- They are ordered, unchangeable(Immutable) and allow duplicates
 """
pop_per_county = (2000,5899,3300,74748,646438, 2002, 2000, 2001)
pop_per_county[0] # First Item
```

```python
# Adding an Item in a tuple
ed_levels = ("Primary", "High School", "University")
to_list = list(ed_levels)
to_list.append("Post Grad")
to_list.insert(0,"Kindergarten")
ed_levels = tuple(to_list)
ed_levels
```

```python
# Combine Two Tuples
program_a = ("Data Science","Data Engineering")
program_b = ("Business Analysis",)
program_a += program_b
all_programs = program_a 
all_programs
```

```python
# Remove an element in a tuple
shoe_brands = ("Balenciaga","Nike","Timberland","Puma")
to_list = list(shoe_brands)
to_list.remove("Puma")
shoe_brands = tuple(to_list)
shoe_brands
```

```python
# Looping Through Items
for brand in range(len(shoe_brands)):
    print(brand, shoe_brands[brand])
```

```python
# Using the enumerate Function
for index, value in enumerate(shoe_brands):
    print(index,value)
```
