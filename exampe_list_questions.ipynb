```python
# Question 4
"""
A restaurant tracks orders in a list:
orders = ["Burger", "Pizza", "Fries", "Soda"]
Remove the first order (served)
Add a new order
Print all remaining orders using a loop"""
orders = ["Burger", "Pizza", "Fries", "Soda"]
del orders[0]
orders
orders.append("Coffee")
for order in orders:
    print(order)
```

```python
# Remove Duplicate Items
orders = ["Burger", "Pizza", "Fries", "Soda","Soda","Fries","Burger","Pizza"]
for order in orders:
    if "Burger" in order :
      orders.remove("Burger")     
      print(orders)
```

```python
# Question 2
"""
You have a list of student marks:

marks = [67, 45, 89, 90, 56, 72]
Find the highest and lowest marks
Add a new student’s mark
Remove a failed mark (below 50)"""

marks = [67, 45, 89, 90, 56, 72]

highest_mark = max(marks)
highest_mark
```

```python
lowest_mark = min(marks)
lowest_mark
```

```python
# Add new Student Marks
marks.append(61)
marks
```

```python
# Remove marks below 50
for mark in marks:
    if mark < 50:
        marks.remove(mark)
marks
```
